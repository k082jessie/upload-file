```python
import os, sys
import cv2
from PIL import Image

os.chdir('/home/nfs/user/jessieyen/PycharmProjects/fashion-compatibility/data/polyvore_outfits')
path = os.getcwd()
print(path)

def make_square(im, min_size=256, fill_color=(0, 0, 0, 0)):
    x, y = im.size
    size = max(min_size, x, y)
    new_im = Image.new('RGBA', (size, size), fill_color)
    new_im.paste(im, (int((size - x) / 2), int((size - y) / 2)))
    return new_im

def image_resize(image, width = None, height = None, inter = cv2.INTER_AREA):
    # initialize the dimensions of the image to be resized and
    # grab the image size
    dim = None
    (h, w) = image.shape[:2]

    # if both the width and height are None, then return the
    # original image
    if width is None and height is None:
        return image

    # check to see if the width is None
    if width is None:
        # calculate the ratio of the height and construct the
        # dimensions
        r = height / float(h)
        dim = (int(w * r), height)

    # otherwise, the height is None
    else:
        # calculate the ratio of the width and construct the
        # dimensions
        r = width / float(w)
        dim = (width, int(h * r))

    # resize the image
    resized = cv2.resize(image, dim, interpolation = inter)

    # return the resized image
    return resized



# 188621378 43654883 158846937 197575483
# 85145844 104626964 105897534 113587483 158091940 159084438 
# 173773682 184092890 191331577 194677737
test_image = Image.open('194677737.jpg')
new_image = make_square(test_image, min_size=300, fill_color=(255, 255, 255, 0))
#new_image.show()
new_image = new_image.convert("RGB")
new_image.save("tmp.jpg")

infile = "tmp.jpg"
im = cv2.imread(infile)
im = image_resize(im, height = 300)
cv2.imwrite("194677737_300.jpg", im)
```
